## 1. Conditionals

### 1.1. `if`, `else` and `elif`

```python
num = 0

if num > 0:
    print('Greater than zero')

elif num == 0:
    print('Equal to zero') 

else:
    print('Not greater than zero')

    
print('This runs every time')
```

## 2. Loops

### 2.1. `for` loop

```python
my_list = ['Python', 'is', 'fun']
```

```python
for word in my_list:
    print(word)
```

### 2.2. `while` loop

```python
count = 0

while count < 5:
    print(f'Count is {count}')
    count += 1
    
print(f'Count is {count}')
```
