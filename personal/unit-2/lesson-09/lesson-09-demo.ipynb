````python
arr_2d = np.array([
    [10, 20, 30, 40],
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])
print(arr_2d)
````

````python
print(arr_2d[0])
````

### 5.3. Filtering

````python
a = np.array([10,20,30,40,50,60,70])
print(a)
````

````python
b = 50
print(b < 60)
````

````python
mask = a > 20
print(mask)
````

````python
print(a[mask])
````

````python
print(a[a > 20])
````

## 6. A note about copies vs 'views'

````python
copy_a = a[2:5].copy()
print(copy_a)
````

````python
copy_a[1] = 100
print(copy_a)
````

````python
print(a)
````

````python
b = a[2:5]
````

````python
b[1] = 100
````

````python
print(b)
````

````python
print(a)
````

````python
id(a)
````

````python
id(b)
````
