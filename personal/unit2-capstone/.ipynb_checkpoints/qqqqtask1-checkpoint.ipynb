```python
#check for duplicates in the dataset
df.duplicated().sum()
#if duplicates are found, remove them
df = df.drop_duplicates()
#if missing values are found, set them to null
df = df.fillna(np.nan)
```

```python
#check unique values in each column
df['employed'].unique()
df['health'].unique()
df['adl'].unique()
df['region'].unique()
df['gender'].unique()
df['married'].unique()
df['school'].unique()
df['income'].unique()
df['employed'].unique()
df['insurance'].unique()
df['medicaid'].unique()
```

```python
#checl value counts in each column
df['employed'].value_counts()
df['health'].value_counts()
df['adl'].value_counts()
df['region'].value_counts()
df['gender'].value_counts()
df['married'].value_counts()
df['school'].value_counts()
df['income'].value_counts()
df['employed'].value_counts()
df['insurance'].value_counts()
df['medicaid'].value_counts()
```

```python
# Comment on any findings from the above steps:
# Data types are string for some columns like employed, insurance, and medicaid. These should be converted to boolean values for easier analysis and optimization.
# Which will save memory and make it easier to work with the data in future. 
```

```python
#change the data type of employed column to boolean
df['employed'] = df['employed'].map({'yes': True, 'no': False})
#change the data type of the insurance column to boolean
df['insurance'] = df['insurance'].map({'yes': True, 'no': False})
#change the data type of the medicaid column to boolean
df['medicaid'] = df['medicaid'].map({'yes': True, 'no': False})
#change the data type of the visit column to int32
df['visits'] = df['visits'].astype(np.int32)
#change the data type of the nvisit column to int32
df['nvisits'] = df['nvisits'].astype(np.int32)
#change the data type of the ovisit column to int32
df['ovisits'] = df['ovisits'].astype(np.int32)
#change the data type of the novisits column to int32
df['novisits'] = df['novisits'].astype(np.int32)
#chnage the data type of the emergency column to int32
df['emergency'] = df['emergency'].astype(np.int32)
#change the data type of the hospital column to int32
df['hospital'] = df['hospital'].astype(np.int32)
#change the data type of the chronic column to int32
df['chronic'] = df['chronic'].astype(np.int32)
#change the data type of the age column to float32
df['age'] = df['age'].astype(np.float32)
#change the data type of the school column to int32
df['school'] = df['school'].astype(np.int32)
#change the data type of the incomme to float32
df['income'] = df['income'].astype(np.float32)
#view info about the dataset again to confirm changes
df.info()
df.describe()
```

```python
# I was trying to save int64 to int32 and float64 to float32 to save memory but after saving to csv and loading it again, the data types are back to int64 and float64.
```

```python
#export the cleaned dataset to a new csv file
df.to_csv('NSMES1988new.csv' , index=False)
```

```python
#load the cleaned dataset into a new dataframe to confirm changes
df_new = pd.read_csv('NSMES1988new.csv')
#view info about the new dataset to confirm changes
df_new.info()
df_new.head()
```

### Task 1 Summary: Data Cleaning and Optimization
1. **Data Quality**: The dataset was inspected for missing values and duplicates. None were found, indicating the data is clean and ready for analysis.
2. **Memory Optimization**: The data types of several columns were updated (e.g., converting 'yes'/'no' strings to booleans and optimizing numeric types). This reduces memory usage and improves analytical performance.
3. **Data Export**: The cleaned and optimized dataset was successfully exported to a new CSV file (NSMES1988new.csv) for future use without having to repeat the cleaning process.

# Starting Task 2

## Task 2: Data visualization

1. Import visualization libraries (matplotlib, seaborn)
2. Select appropriate visualization techniques for the data:
    - Bar plot for average healthcare visits by Health
    - Bar plot for healthcare visits by Region
    - Scatter plot/heatmap showing correlation among numerical variables
    - Box plot to identify outliers in Age, Income, and Visits
3. Record observations and insights from visualizations
4. Save plots and observations for reporting purposes

```python
# Import visualization libraries (matplotlib, seaborn)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar plot for average healthcare visits by Health
plt.title('Average Healthcare Visits by Health')
sns.barplot(data=df_new, x='health', y='visits', color='orange', edgecolor='black', errorbar=None, label='Average Healthcare Visits')
plt.xlabel('Health')
plt.ylabel('Healthcare Visits')
plt.legend()
plt.show()
```

```python
# Bar plot for healthcare visits by Region
plt.title('Healthcare Visits by Region')
sns.barplot(data=df_new, x='region', y='visits', color='blue', edgecolor='black', errorbar=None, label='Healthcare Visits')
plt.xlabel('Region')
plt.ylabel('Healthcare Visits')
plt.legend()
plt.show()
```

```python
# Scatter plot/heatmap showing correlation among numerical variables
plt.figure(figsize=(10, 8))
numeric_cols = df_new.select_dtypes(include=['number', 'bool']).columns
sns.heatmap(df_new[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()

plt.figure(figsize=(10, 8))
plt.scatter(df_new['age'], df_new['visits'], color='green', edgecolor='black', label='Age vs Visits')
plt.xlabel('Age')
plt.ylabel('Healthcare Visits')
plt.title('Scatter Plot of Age vs Healthcare Visits')
plt.legend()
plt.show()
```

```python
# Box plot to identify outliers in Age, Income, and Visits
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_new[['age', 'income', 'visits']], patch_artist=True, notch=True)
plt.title('Box Plot of Age, Income, and Healthcare Visits')
plt.show()
```

## Task 2 Observations and Insights

Based on the visualizations generated above, we can observe the following:

1. **Average Healthcare Visits by Health**: Patients who self-reported having 'poor' health have a noticeably higher average number of healthcare visits compared to those with 'average' or 'excellent' health.
2. **Healthcare Visits by Region**: The distribution of visits across regions shows how healthcare utilization is spread geographically, though health status appears to be a much stronger indicator for visits than location.
3. **Correlation Heatmap**: There is a logical positive correlation between visits and health-related metrics like chronic conditions and hospital stays.
4. **Age vs. Visits Scatter Plot**: The scatter plot shows a dense cluster of patients with lower visit counts across all ages. However, it also reveals that extreme high-visit cases occur across the entire senior age spectrum (66+).
5. **Outliers (Box Plots)**: The box plots clearly confirm the presence of significant outliers. Income is heavily right-skewed with massive extreme values, and visits also contains distinct upper outliers (e.g., a few patients requiring a very high number of visits).

## Task 3: Data processing and statistical analysis

1. Perform transformations:
    - Multiply Age by 10
    - Multiply Income by 10,000
2. Conduct basic statistical analysis:
    - Use describe() function for key columns like Age, Income, and Visits
    - Compare the results with raw dataset statistics
    - Identify columns that are not suitable for statistical analysis and recommend possible datatype changes
3. Export the processed data to a CSV file named NSMES1988updated.csv
4. Prepare a short report on statistical observations and insights

```python
# Task 3: Data processing and statistical analysis
# Perform transformations
df_new['age'] = df_new['age'] * 10
df_new['income'] = df_new['income'] * 10000
```

```python
# Task 3: Data processing and statistical analysis
# Conduct basic statistical analysis on the transformed data
df_new.describe()
```

```python
# Identify columns not suitable for statistical analysis
# 1. 'Unnamed: 0' is just an ID/index column and holds no statistical value.
# 2. 'health', 'adl', 'region', 'gender', and 'married' are nominal/ordinal strings.
# 3. Convert these string columns to 'category' datatype.
categorical_cols = ['health', 'adl', 'region', 'gender', 'married']
df_new[categorical_cols] = df_new[categorical_cols].astype('category')
df_new = df_new.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
# Export the processed data to a CSV file named NSMES1988updated.csv
df_new.to_csv('NSMES1988updated.csv', index=False)
print("Processed data exported to NSMES1988updated.csv successfully!")
```

### Task 3 Report: Statistical Observations and Insights
1. **Age Analysis**: After multiplying by 10, the mean age is ~74.02 years. The distribution spans from 66 to 109 years, confirming this is a senior demographic.
2. **Income Analysis**: The mean income is ~$25,271. The standard deviation is huge (~$29,246) compared to the median ($16,981), indicating a highly right-skewed income distribution with massive outliers (max $548,351).
3. **Healthcare Visits**: The average patient has about 5.77 visits. The median is 4, but the maximum is 89, indicating that a small group of patients requires significantly more medical attention.
4. **Datatype Updates**: String-based categorical columns were converted to category datatypes. This optimizes memory and prepares the dataset for proper statistical groupings.
