```python
#change the data type of employed column to boolean
df['employed'] = df['employed'].map({'yes': True, 'no': False})
#change the data type of the insurance column to boolean
df['insurance'] = df['insurance'].map({'yes': True, 'no': False})
#change the data type of the medicaid column to boolean
df['medicaid'] = df['medicaid'].map({'yes': True, 'no': False})
#change the data type of the visit column to int32
df['visits'] = df['visits'].astype(np.int32)
#change the data type of the nvisit column to int32
df['nvisits'] = df['nvisits'].astype(np.int32)
#change the data type of the ovisit column to int32
df['ovisits'] = df['ovisits'].astype(np.int32)
#change the data type of the novisits column to int32
df['novisits'] = df['novisits'].astype(np.int32)
#chnage the data type of the emergency column to int32
df['emergency'] = df['emergency'].astype(np.int32)
#change the data type of the hospital column to int32
df['hospital'] = df['hospital'].astype(np.int32)
#change the data type of the chronic column to int32
df['chronic'] = df['chronic'].astype(np.int32)
#change the data type of the age column to float32
df['age'] = df['age'].astype(np.float32)
#change the data type of the school column to int32
df['school'] = df['school'].astype(np.int32)
#change the data type of the incomme to float32
df['income'] = df['income'].astype(np.float32)
#view info about the dataset again to confirm changes
df.info()
df.describe()
```

## Task 4: Data analysis with Pandas

1. Identify categorical and numerical variables
2. Focus on columns such as Health and Region
3. Perform pivoting operations on the dataset based on categorical columns:
4. Group by Health and calculate the average healthcare visits
5. Analyze visit trends across Region
6. Create distribution tables:
   - Healthcare visits distribution by Health
   - Healthcare visits distribution by Region
7. Encode categorical variables and save data as NSMES1988_Dummy.csv


```python
# Step 1: Identify categorical and numerical variables
categorical_cols = df_new.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
numerical_cols = df_new.select_dtypes(include=['number']).columns.tolist()

print('Categorical variables:', categorical_cols)
print('Numerical variables:', numerical_cols)
```

```python
# Steps 3 & 4: Pivoting / grouping operations on categorical columns
# Group by Health and calculate the average healthcare visits
avg_visits_by_health = df_new.groupby('health', observed=True)['visits'].mean().sort_values(ascending=False)
print('Average healthcare visits by Health:')
print(avg_visits_by_health)

# Pivot table: average visits broken down by both Health and Region
pivot_health_region = pd.pivot_table(
    df_new,
    values='visits',
    index='health',
    columns='region',
    aggfunc='mean',
    observed=True
)
print('\nPivot table (average visits by Health and Region):')
print(pivot_health_region)
```

```python
# Step 5: Analyze visit trends across Region
visit_trends_by_region = df_new.groupby('region', observed=True)['visits'].agg(['mean', 'median', 'min', 'max', 'count'])
print('Healthcare visit trends across Region:')
print(visit_trends_by_region)
```

```python
# Step 6: Create distribution tables
# Total and average healthcare visits distribution by Health
visits_dist_health = df_new.groupby('health', observed=True)['visits'].agg(
    total_visits='sum', average_visits='mean', patient_count='count'
)
print('Healthcare visits distribution by Health:')
print(visits_dist_health)

# Total and average healthcare visits distribution by Region
visits_dist_region = df_new.groupby('region', observed=True)['visits'].agg(
    total_visits='sum', average_visits='mean', patient_count='count'
)
print('\nHealthcare visits distribution by Region:')
print(visits_dist_region)
```

```python
# Step 7: Encode categorical variables and save as NSMES1988_Dummy.csv
# One-hot encode the nominal/ordinal string columns (drop_first avoids the
# dummy-variable trap). The boolean columns (employed/insurance/medicaid) are
# already 0/1-style, so they don't need encoding.
cols_to_encode = ['health', 'adl', 'region', 'gender', 'married']
df_dummy = pd.get_dummies(df_new, columns=cols_to_encode, drop_first=True)

print('Shape after encoding:', df_dummy.shape)
print('New encoded columns:')
print([c for c in df_dummy.columns if any(c.startswith(p + '_') for p in cols_to_encode)])

# Export the encoded dataset
df_dummy.to_csv('NSMES1988_Dummy.csv', index=False)
print('\nEncoded data exported to NSMES1988_Dummy.csv successfully!')
df_dummy.head()
```

### Task 4 Report: Pandas Data Analysis Observations
1. **Health is the strongest driver of visits**: Grouping by `health` shows average visits rise sharply as health declines — from ~3.4 (excellent) to ~5.5 (average) to ~8.9 (poor). This is the clearest pattern in the dataset.
2. **Region has a secondary effect**: Average visits by region are fairly close (~5.4 midwest to ~6.4 west), confirming that *health status* matters far more than *location*. The pivot table shows the health → visits trend holds within every region.
3. **Distribution tables**: The `average` health group contains the large majority of patients, but the smaller `poor` group accounts for a disproportionate share of total visits, reflecting heavier healthcare utilization by the sickest patients.
4. **Encoding**: Nominal/ordinal string columns (`health`, `adl`, `region`, `gender`, `married`) were one-hot encoded with `drop_first=True` to avoid the dummy-variable trap. The already-boolean columns (`employed`, `insurance`, `medicaid`) were left as-is. The fully numeric, model-ready dataset was exported to `NSMES1988_Dummy.csv`.
